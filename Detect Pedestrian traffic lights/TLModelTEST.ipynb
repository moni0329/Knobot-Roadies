Test the Traffic Light Detection Model

```python
import torch
from google.colab import drive
from PIL import Image
from IPython.display import display
import os

repo_path = '/content/yolov5'
model_path = '/content/drive/MyDrive/Models/Best models TL/best.pt'
img_path = '/content/drive/MyDrive/TestImg/tl_test_img/red1.jpg'

model = torch.hub.load(repo_path, 'custom', path=model_path, source='local')

model.eval()

# Perform inference on the test image
results = model(img_path)

results.print()

save_dir = '/content/yolov5/runs/detect'
results.save(save_dir=save_dir)

# Get the path to the saved image
processed_image_path = os.path.join(save_dir, 'exp', results.files[0])


# Analyze the results
for pred in results.xyxy[0]:  # Iterate through detections
    class_id = int(pred[5])  # Class ID
    confidence = float(pred[4])  # Confidence score
    if class_id == 0:
        print(f"Detected: Green Light with confidence {confidence:.2f}")
    elif class_id == 2:
        print(f"Detected: Red Light with confidence {confidence:.2f}")
```

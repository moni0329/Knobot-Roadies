Test the Crosswalk Detection Model

```python
import torch

# Path to YOLOv5 repository
repo_path = '/content/drive/MyDrive/yolov5'

# Load the trained YOLOv5 model
model = torch.hub.load(repo_path, 'custom', path='/content/drive/MyDrive/Models/BestmodelsCW/best.pt', source='local')

# Set the model to evaluation mode
model.eval()

# Example of using the model for inference
img = '/content/drive/MyDrive/TestImg/cw_test_img/crosswalk_image124.png'

# Perform inference
results = model(img)

# Print results
results.print()  # Print results to the console

# Save results to an image file
results.save('/content/drive/MyDrive/TestImg')
# To access the bounding boxes and other details
print(results.xyxy)  # Bounding box coordinates and scores
```
